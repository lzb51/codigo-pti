```python
def coletar_informacoes_caixa(numero_caixas):
    informacoes_caixas = []

    for i in range(numero_caixas):
        print(f"\nCaixa {i+1}:")
        altura = float(input("Qual é a altura da caixa em centímetros? "))
        largura = float(input("Qual é a largura da caixa em centímetros? "))
        profundidade = float(input("Qual é a profundidade da caixa em centímetros? "))
        peso = float(input("Qual é o peso da caixa em quilogramas? "))

        print("\nEm uma escala de 1 a 5, o quão frágil é a caixa?")
        while True:
            fragilidade = int(input("Digite um número de 1 a 5 (sendo 1 menos frágil e 5 mais frágil): "))
            if 1 <= fragilidade <= 5:
                break
            else:
                print("Valor inválido! Digite um número entre 1 e 5.")

        informacoes_caixa = {
            "altura": altura,
            "largura": largura,
            "profundidade": profundidade,
            "peso": peso,
            "fragilidade": fragilidade
        }

        informacoes_caixas.append(informacoes_caixa)

    return informacoes_caixas

quantidade_caixas = int(input("Quantas caixas você gostaria de inserir? "))
caixas = coletar_informacoes_caixa(quantidade_caixas)

print("\nInformações das caixas:")
for i, caixa in enumerate(caixas):
    print(f"\nCaixa {i+1}:")
    print(f"Altura da caixa: {caixa['altura']} cm")
    print(f"Largura da caixa: {caixa['largura']} cm")
    print(f"Profundidade da caixa: {caixa['profundidade']} cm")
    print(f"Peso da caixa: {caixa['peso']} kg")
    print(f"Fragilidade da caixa: {caixa['fragilidade']} (sendo 1 menos frágil e 5 mais frágil)")
```

```python
from py3dbp import Packer, Bin, Item

packer = Packer()

packer.add_bin(Bin('small-envelope', 11.5, 6.125, 0.25, 10))
packer.add_bin(Bin('large-envelope', 15.0, 12.0, 0.75, 15))
packer.add_bin(Bin('small-box', 8.625, 5.375, 1.625, 70.0))
packer.add_bin(Bin('medium-box', 11.0, 8.5, 5.5, 70.0))
packer.add_bin(Bin('medium-2-box', 13.625, 11.875, 3.375, 70.0))
packer.add_bin(Bin('large-box', 12.0, 12.0, 5.5, 70.0))
packer.add_bin(Bin('large-2-box', 23.6875, 11.75, 3.0, 70.0))

packer.add_item(Item('50g [item 1]', 3.9370, 1.9685, 1.9685, 1))
packer.add_item(Item('50g [item 2]', 3.9370, 1.9685, 1.9685, 2))
packer.add_item(Item('50g [item 3]', 3.9370, 1.9685, 1.9685, 3))
packer.add_item(Item('250g [item 4]', 7.8740, 3.9370, 1.9685, 4))
packer.add_item(Item('250g [item 5]', 7.8740, 3.9370, 1.9685, 5))
packer.add_item(Item('250g [item 6]', 7.8740, 3.9370, 1.9685, 6))
packer.add_item(Item('250g [item 7]', 7.8740, 3.9370, 1.9685, 7))
packer.add_item(Item('250g [item 8]', 7.8740, 3.9370, 1.9685, 8))
packer.add_item(Item('250g [item 9]', 7.8740, 3.9370, 1.9685, 9))

packer.pack()

for b in packer.bins:
    print(":::::::::::", b.string())

    print("FITTED ITEMS:")
    for item in b.items:
        print("====> ", item.string())

    print("UNFITTED ITEMS:")
    for item in b.unfitted_items:
        print("====> ", item.string())

    print("***************************************************")
    print("***************************************************")
```

```python
import plotly.graph_objects as go

# Dimensões da caixa azul
x_length_blue = 11000
y_length_blue = 8500
z_length_blue = 5500

# Pontos da caixa azul
x_blue = [0, 0, x_length_blue, x_length_blue,
          0, 0, x_length_blue, x_length_blue]
y_blue = [0, y_length_blue, y_length_blue, 0,
          0, y_length_blue, y_length_blue, 0]
z_blue = [0, 0, 0, 0,
          z_length_blue, z_length_blue, z_length_blue, z_length_blue]

# Definição das arestas da caixa azul
edges_blue = [
    [0, 1], [1, 2], [2, 3], [3, 0],  # Arestas da face inferior
    [4, 5], [5, 6], [6, 7], [7, 4],  # Arestas da face superior
    [0, 4], [1, 5], [2, 6], [3, 7],  # Arestas verticais
    [4, 6], [5, 7]                   # Arestas horizontais
]

# Criação do primeiro gráfico 3D interativo com a caixa azul
fig_blue = go.Figure()

# Adição das arestas da caixa azul
for edge in edges_blue:
    x_edge = [x_blue[edge[0]], x_blue[edge[1]]]
    y_edge = [y_blue[edge[0]], y_blue[edge[1]]]
    z_edge = [z_blue[edge[0]], z_blue[edge[1]]]
    fig_blue.add_trace(go.Scatter3d(
        x=x_edge,
        y=y_edge,
        z=z_edge,
        mode='lines',
        line=dict(color='blue', width=3)
    ))

# Configuração dos rótulos dos eixos para o primeiro gráfico
max_dim_blue = max(x_length_blue, y_length_blue, z_length_blue)
fig_blue.update_layout(
    scene=dict(
        xaxis=dict(title='Eixo X', range=[-max_dim_blue, x_length_blue]),
        yaxis=dict(title='Eixo Y', range=[-max_dim_blue, y_length_blue]),
        zaxis=dict(title='Eixo Z', range=[-max_dim_blue, z_length_blue]),
        aspectmode='cube'
    )
)

# Exibição do primeiro gráfico interativo
fig_blue.show()

# Dimensões da caixa amarela
x_length_yellow = 3937
y_length_yellow = 1968
z_length_yellow = 1968

# Posição da caixa amarela
x_start_yellow = x_length_blue + 500  # Adjust this value for spacing between blue and yellow boxes

# Pontos da caixa amarela
x_yellow = [x_start_yellow, x_start_yellow, x_start_yellow + x_length_yellow, x_start_yellow + x_length_yellow,
            x_start_yellow, x_start_yellow, x_start_yellow + x_length_yellow, x_start_yellow + x_length_yellow]
y_yellow = [0, y_length_yellow, y_length_yellow, 0,
            0, y_length_yellow, y_length_yellow, 0]
z_yellow = [0, 0, 0, 0,
            z_length_yellow, z_length_yellow, z_length_yellow, z_length_yellow]

# Definição das arestas da caixa amarela
edges_yellow = [
    [0, 1], [1, 2], [2, 3], [3, 0],  # Arestas da face inferior
    [4, 5], [5, 6], [6, 7], [7, 4],  # Arestas da face superior
    [0, 4], [1, 5], [2, 6], [3, 7],  # Arestas verticais
    [4, 6], [5, 7]                   # Arestas horizontais
]

# Criação do segundo gráfico 3D interativo com as duas caixas e suas arestas
fig_both = go.Figure()

# Adição das arestas da caixa azul
for edge in edges_blue:
    x_edge = [x_blue[edge[0]], x_blue[edge[1]]]
    y_edge = [y_blue[edge[0]], y_blue[edge[1]]]
    z_edge = [z_blue[edge[0]], z_blue[edge[1]]]
    fig_both.add_trace(go.Scatter3d(
        x=x_edge,
        y=y_edge,
        z=z_edge,
        mode='lines',
        line=dict(color='blue', width=3)
    ))

# Adição das arestas da caixa amarela
for edge in edges_yellow:
    x_edge = [x_yellow[edge[0]], x_yellow[edge[1]]]
    y_edge = [y_yellow[edge[0]], y_yellow[edge[1]]]
    z_edge = [z_yellow[edge[0]], z_yellow[edge[1]]]
    fig_both.add_trace(go.Scatter3d(
        x=x_edge,
        y=y_edge,
        z=z_edge,
        mode='lines',
        line=dict(color='yellow', width=3)
    ))

# Configuração dos rótulos dos eixos para o segundo gráfico
max_dim_both = max(x_start_yellow + x_length_yellow, y_length_blue, max(z_length_blue, z_length_yellow))
fig_both.update_layout(
    scene=dict(
        xaxis=dict(title='Eixo X', range=[-max_dim_both, x_start_yellow + x_length_yellow]),
        yaxis=dict(title='Eixo Y', range=[-max_dim_both, y_length_blue]),
        zaxis=dict(title='Eixo Z', range=[-max_dim_both, max(z_length_blue, z_length_yellow)]),
        aspectmode='cube'
    )
)

# Exibição do segundo gráfico interativo com as duas caixas
fig_both.show()

# Dimensões da caixa verde
x_length_green = x_length_yellow
y_length_green = y_length_yellow
z_length_green = z_length_blue

# Posição da caixa verde
x_start_green = x_start_yellow + x_length_yellow + 500  # Adjust this value for spacing between yellow and green boxes

# Pontos da caixa verde
x_green = [x_start_green, x_start_green, x_start_green + x_length_green, x_start_green + x_length_green,
           x_start_green, x_start_green, x_start_green + x_length_green, x_start_green + x_length_green]
y_green = [0, y_length_green, y_length_green, 0,
           0, y_length_green, y_length_green, 0]
z_green = [0, 0, 0, 0,
           z_length_green, z_length_green, z_length_green, z_length_green]

# Definição das arestas da caixa verde
edges_green = [
    [0, 1], [1, 2], [2, 3], [3, 0],  # Arestas da face inferior
    [4, 5], [5, 6], [6, 7], [7, 4],  # Arestas da face superior
    [0, 4], [1, 5], [2, 6], [3, 7],  # Arestas verticais
    [4, 6], [5, 7]                   # Arestas horizontais
]

# Criação do terceiro gráfico 3D interativo com as três caixas e suas arestas
fig_all = go.Figure()

# Adição das arestas da caixa azul
for edge in edges_blue:
    x_edge = [x_blue[edge[0]], x_blue[edge[1]]]
    y_edge = [y_blue[edge[0]], y_blue[edge[1]]]
    z_edge = [z_blue[edge[0]], z_blue[edge[1]]]
    fig_all.add_trace(go.Scatter3d(
        x=x_edge,
        y=y_edge,
        z=z_edge,
        mode='lines',
        line=dict(color='blue', width=3)
    ))

# Adição das arestas da caixa amarela
for edge in edges_yellow:
    x_edge = [x_yellow[edge[0]], x_yellow[edge[1]]]
    y_edge = [y_yellow[edge[0]], y_yellow[edge[1]]]
    z_edge = [z_yellow[edge[0]], z_yellow[edge[1]]]
    fig_all.add_trace(go.Scatter3d(
        x=x_edge,
        y=y_edge,
        z=z_edge,
        mode='lines',
        line=dict(color='yellow', width=3)
    ))

# Adição das arestas da caixa verde
for edge in edges_green:
    x_edge = [x_green[edge[0]], x_green[edge[1]]]
    y_edge = [y_green[edge[0]], y_green[edge[1]]]
    z_edge = [z_green[edge[0]], z_green[edge[1]]]
    fig_all.add_trace(go.Scatter3d(
        x=x_edge,
        y=y_edge,
        z=z_edge,
        mode='lines',
        line=dict(color='green', width=3)
    ))

# Configuração dos rótulos dos eixos para o terceiro gráfico
max_dim_all = max(x_start_green + x_length_green, y_length_blue, max(z_length_blue, z_length_green))
fig_all.update_layout(
    scene=dict(
        xaxis=dict(title='Eixo X', range=[-max_dim_all, x_start_green + x_length_green]),
        yaxis=dict(title='Eixo Y', range=[-max_dim_all, y_length_blue]),
        zaxis=dict(title='Eixo Z', range=[-max_dim_all, max(z_length_blue, z_length_green)]),
        aspectmode='cube'
    )
)

# Exibição do terceiro gráfico interativo com as três caixas
fig_all.show()
```
